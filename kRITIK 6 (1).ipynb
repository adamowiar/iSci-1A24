```python
import matplotlib.pyplot as plt
import numpy as np

def normal_density(mean, variance, x):
    coeff = 1/np.sqrt(variance*2*np.pi)
    exp_term = np.exp(-(x-mean)**2 / (2*variance))
    return coeff*exp_term
def integrating(mean, variance, a, b, num_steps=90000):
    d_x = (b-a)/num_steps
    x_values = np.linspace(a,b, num_steps)
    areaofgraph = np.sum(normal_density(mean, variance, x_values)*d_x)
    return areaofgraph
def distribution_plot(mean, variance, a, b, num_points= 90000): #GREATER NUMBER MEANS SMOOTHER
    x_spot = np.linspace(a,b,num_points)
    y_spot = normal_density (mean, variance, x_spot) #varianceto must be included as we can't know for sure what is going to happen, basically impossible to do so 

    plt.plot(x_spot, y_spot)
    plt.show()

result = integrating(171, (7.1**2), 162, 190)
distribution_plot(171, (7.1**2), 162, 190)

peint("Integrated area", result)




```
